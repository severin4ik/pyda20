# Работа с файловой системой и модули


## 1. 
Переведите содержимое файла purchase_log.txt в словарь purchases вида: {‘1840e0b9d4’: ‘Продукты’, …}

```python
purchases = {}
```

```python
import json 

with open('purchase_log.txt') as f:
    f.readline() #не получилось иначе начать чтение со второй строки, поставил курсор на вторую.
    for line in f:
        line = line.strip()
        
        dict_ = json.loads(line)
        
        purchases[dict_['user_id']] = dict_['category']
```

### 2.  
Для каждого user_id в файле visit_log.csv определите третий столбец с категорией покупки (если покупка была, сам файл visit_log.csv изменять не надо). Запишите в файл funnel.csv визиты из файла visit_log.csv, в которых были покупки с указанием категории.
    
Учтите условия на данные:
- содержимое purchase_log.txt помещается в оперативную память компьютера
- содержимое visit_log.csv - нет; используйте только построчную обработку этого файла


```python
import json

with open('visit_log.csv', 'r') as vl:
    with open('funnel.csv', 'w') as f2write:
        f2write.write("user_id,source,category\n") #вручную пишем заголовок
        
#Построчно проходимся по visit_log, приведя строку в список: 
#line[0] - userid, line[1] - source откуда пришли. 
        for i, line in enumerate(vl):
            if i > 0: #пропускаем заголовок
                line = line.strip()
                line = line.strip().split(',') 
                
#Проверяем, если user_id из visit_log находится в purchase_log 
#(,кот. преобразован в словарь purchases[user_id] = category): 
            if line[0] in purchases:
                new_record = (line[0],line[1], purchases[line[0]],"\n")
                new_record = ",".join(new_record)
                f2write.write(new_record) #записываем результат в f2write
            
```
