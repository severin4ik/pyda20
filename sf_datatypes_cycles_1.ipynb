## Задание 1

    среднюю букву, если число букв в слове нечетное;

```python
word = "testing"

a = len(word) #функция количества знаков в строке

if len(word)%2 == 0: #проверяем на четность
    print(word[int(a/2)-1],word[int(a/2)]) #делим пополам, нужная по счету буква, но т.к. индекс начинается с нуля, идет сдвиг на 1
else:
    print(word[int(a//2)]) #целочисленное деление, указывает на нужный индекс
    
```

## Задание 5 (необязательное)

Дан поток логов по количеству просмотренных страниц для каждого пользователя. Вам необходимо написать алгоритм, который считает среднее значение просмотров на пользователя. Т. е. надо посчитать отношение суммы всех просмотров к количеству уникальных пользователей.

```python
stream = [
    '2018-01-01,user1,3',
    '2018-01-07,user1,4',
    '2018-03-29,user1,1',
    '2018-04-04,user1,13',
    '2018-01-05,user2,7',
    '2018-06-14,user3,4',
    '2018-07-02,user3,10',
    '2018-03-21,user4,19',
    '2018-03-22,user4,4',
    '2018-04-22,user4,8',
    '2018-05-03,user4,9',
    '2018-05-11,user4,11',
]


unique_users = []
rating_sum = 0
# преобразование строки в список.
for string in stream:
    x = stream.index(string)
    stream[x] = string.split(',') #NB! не пойму почему не получается string = string.split(',')

#Итерация списка списков
for record in stream:
    if record[1] not in unique_users:
        unique_users.append(record[1]) #составляем список уникальных пользователей
    rating_sum += int(record[2]) #суммируем кол-во просмотров
        

print(f"Среднее количество просмотров на уникального пользователя: {round(rating_sum/len(unique_users),2)}")
```

## Задание 6 (необязательное)

Необходимо у пользователя запрашивать набор чисел разделенных пробелом. В результате в отсортированном порядке должны выводиться числа, которые повторяются в вводе более одного раза.

```python
numbers_string = input("Введите числа, разделенные пробелом): ")
numbers_list = numbers_string.split(' ') 

repetitive_numbers = []

for number in numbers_list:
    if numbers_list.count(number) > 1 and int(number) not in repetitive_numbers: #проверяем есть ли строка в списке, а добавляем преобразованный тип
        repetitive_numbers.append(int(number))

repetitive_numbers = sorted(repetitive_numbers)

if len(repetitive_numbers) != 0:
    for number in repetitive_numbers:
        print(number, end = ' ')
    
else:
    print("Числа не повторяются!")
```
