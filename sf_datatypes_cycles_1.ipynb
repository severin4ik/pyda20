## Задание 1

Дана переменная, в которой хранится слово из латинских букв. Напишите код, который выводит на экран:

    среднюю букву, если число букв в слове нечетное;
    две средних буквы, если число букв четное.


```python
word = "testing"

a = len(word) #функция количества знаков в строке

if len(word)%2 == 0: #проверяем на четность
    print(word[int(a/2)-1],word[int(a/2)]) #делим пополам, нужная по счету буква, но т.к. индекс начинается с нуля, идет сдвиг на 1
else:
    print(word[int(a//2)]) #целочисленное деление, указывает на нужный индекс
    
```

## Задание 2

Напишите программу, которая последовательно запрашивает у пользователя числа (по одному за раз) и после первого нуля выводит сумму всех ранее введенных чисел.

```python
string = []
number = 1
#цикл while повторяет последовательность (запрос числа, запись его в список) до тех пор, пока условие "переменная number не равно 0", истино
while number != 0:
    number = int(input("Введите число:"))
    string.append(number)
print("Результат:" , sum(string))
```
