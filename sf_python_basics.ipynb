## Задание 1

Напишите код, который проверяет какая из этих строк длиннее.

## Задание 5 

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.



```python
## взятием остатка от деления выделяем последнюю цифру
## целочисленным делением, убираем последние цифры

number = 123321
number_1=number//1%10
number_2=number//10%10
number_3=number//100%10
number_4=number//1000%10
number_5=number//10000%10
number_6=number//100000%10

## print(number_1,number_2,number_3,number_4,number_5,number_6)

if (number_1 + number_2 + number_3 == number_4 + number_5 + number_6):
    print ("Результат: Счастливый билет")
else:
    print ("Результат: Обычный билет")
```


## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

    если пользователь выбрал круг, запрашиваем его радиус,
    если треугольник – длины трех его сторон;
    если прямоугольник – длины двух его сторон.


```python
shape_input = str(input("Введите тип фигуры (Круг, Треугольник или Прямоугольник): "))

if shape_input == "Круг" or shape_input == "круг":
    radius_input = int(input("Введите радиус круга: "))
    s_circle = 3.14159292 * float(radius_input**2)
    print ("Площадь круга:", round(s_circle,2))
elif shape_input == "Треугольник" or shape_input == "треугольник":
    side_A = int(input("Введите длину стороны A: "))
    side_B = int(input("Введите длину стороны B: "))
    side_C = int(input("Введите длину стороны C: "))
    
    p = (side_A + side_B + side_C)/2
    s_triangle = (p*(p-side_A)*(p-side_B)*(p-side_C))**(0.5)
   
    print ("Площадь треугольника:", round(s_triangle,2))
    
elif shape_input == "Прямоугольник" or shape_input == "прямоугольник":
    side_A = int(input("Введите длину стороны A: "))
    side_B = int(input("Введите длину стороны B: "))
    print ("Площадь прямоугольника:", side_A*side_B)
else:
    print("Введите правильно тип фигуры(Круг, Треугольник или Прямоугольник)")
```
