Вам нужно помочь секретарю автоматизировать работу. Для этого нужно написать программу, которая будет на основе хранимых данных исполнять пользовательские команды.

```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

```python
# Пользователь по команде "p" может узнать владельца документа по его номеру

def find_owner(doc_nr):

    #number = input("Введите номер документа: ")
    
    for document in documents:
        if document['number'] == doc_nr:
            return f"Владелец документа: {document['name']}"
    else:
        return "Документ не найден в базе"
```

```python
# Пользователь по команде "s" может по номеру документа узнать на какой полке он хранится
def find_shelf(doc_nr):
    
    for shelfs, docs in directories.items():
        if doc_nr in docs:
            return shelfs
    else:
        return 0
```

```python
#Информация по имеющимся полкам
def print_shelfs():
    print('Текущий перечень полок: ', end = ' ')
    for shelfs in directories.keys():
        print(shelfs, end = ' ')

print_shelfs()
```

```python
#Пользователь по команде "l" может увидеть полную информацию по всем документам
def all_info():
    info = ['Текущий список документов: ']
    for docs in documents: 
        print(f"№: {docs['number']}, тип: {docs['type']}, владелец: {docs['name']}, полка хранения: {find_shelf(docs['number'])}")
    return ""

all_info()
```

```python
#Пользователь по команде "ads" может добавить новую полку

def add_shelf(new_shelf):

    if new_shelf in directories.keys():
        print("Такая полка уже существует. ", end = '')
       
    else:
        directories[new_shelf] = []
        print(f"Полка добавлена. ", end = '')
    
    print_shelfs()
    
    return ""
   
    
    
   
```

```python
#Пользователь по команде "ds" может удалить существующую полку из данных (только если она пустая) #.get
def delete_shelf(del_shelf):
    if del_shelf in directories.keys() and len(directories[del_shelf]) == 0:
        del(directories[del_shelf])
    elif del_shelf in directories.keys() and len(directories[del_shelf]) != 0:
        print("На полке есть документы, удалите их перед удалением полки.")
    else: print("Такой полки не существует. ")
        
    print_shelfs()
    
    return ""
    
```

```python
#Пользователь по команде "ad" может добавить новый документ в данные. Немного поменял интеракцию с пользователем. 
#теперь мы проверем наличие дркумета в системе по нахождению на полке. Если 0, то документа нет в системе.
def add_document(doc_nr, shelf_nr):
    if find_shelf(doc_nr) == 0 and shelf_nr in directories.keys(): 
        doc_type = input("Введите тип документа: ")
        doc_owner = input("Введите владельца документа: ")

        directories[shelf_nr].append(doc_nr)
        documents.append({})
        documents[-1]['type'] = doc_type 
        documents[-1]['number'] = doc_nr
        documents[-1]['name'] = doc_owner
        
    if find_shelf(doc_nr) != 0:
        print("Такой документ уже существует.")
        
    if shelf_nr not in directories.keys():
        print("Такой полки не существует. Добавьте полку командой as.")

    all_info()
    
    return ""
```

```python
#Пользователь по команде "d" может удалить документ из данных
def del_document(doc_nr):
    i=0
    for document in documents:
        if document['number'] == doc_nr:
            del(documents[i])
            for docs in directories.values():
                if doc_nr in docs:
                    docs.remove(doc_nr)
            break        
        i += 1
            
    else:
        print("Документ не найден в базе.")
    
    all_info()
    
    return ""
```

```python
#Пользователь по команде "m" может переместить документ с полки на полку
#использую find_shelf, чтобы не делать лишний раз итерацию по values дважды
def move_document(doc_nr, shelf_nr):
       
    if find_shelf(doc_nr) != 0 and shelf_nr in directories.keys():  
        directories[find_shelf(doc_nr)].remove(doc_nr)
        directories[shelf_nr].append(doc_nr)
    elif find_shelf(doc_nr) == 0:
        print("Документ не найден в базе.")
    elif shelf_nr not in directories.keys():
        print("Такой полки не существует.")
    all_info()
    
    return ""
```

```python
#1 - функция принимает 1 параметр номер документа
#0 - функция принимает 1 параметр номер полки
#2 - функция принимает 2 параметра, номер документа и номер полки
#find_shelf отдельно, тк функция возвращает только номер полки или 0.

def main():
    command_dict ={
        "p": [find_owner, 1],
        #"s": [find_shelf, 0],
        "l": [all_info], 
        "ads": [add_shelf, 0],
        "ds": [delete_shelf, 0],
        "ad": [add_document, 2],
        "d": [del_document, 1],
        "m": [move_document, 2]
    }
    
    print('''-----------------------------------------------
- вывести полную информацию по документам: [L]
- узнать владельца по номеру документа: [p]
- узнать на какой полке хранится документ: [s]
- добавить новую полку [ads]
- удалить существующую полку [ds]
- добавить новый документ [ad] 
- удалить документ [d]
- переместить документ на новую полку [m]
-----------------------------------------------
          ''')

    while True:
        print('\n')
        user_cmd = (input("Введите комманду: ")).lower()  #не будет ошибки, если больщими буквами введут команду
        
        if user_cmd in command_dict and len(command_dict[user_cmd]) == 1: #команды, функций без вх. параметров
            res = command_dict[user_cmd][0]()
            print(res) 
            
        elif user_cmd in command_dict and command_dict[user_cmd][1] == 1:
            number = input("Введите номер документа: ")
            res = command_dict[user_cmd][0](number)
            print(res)
            
        elif user_cmd in command_dict and command_dict[user_cmd][1] == 0:
            number = input("Введите номер полки: ")
            res = command_dict[user_cmd][0](number)
            print(res)
            
        elif user_cmd in command_dict and command_dict[user_cmd][1] == 2:
            doc_nr = input("Введите номер документа: ")
            shelf_nr = input("Введите номер полки: ") 
            res = command_dict[user_cmd][0](doc_nr, shelf_nr)
            print(res)  
             
        elif user_cmd == "s" or user_cmd == "S":
            number = input("Введите номер документа: ")
            find_shelf(number)
            if find_shelf(number) != 0:
                print(f"Документ хранится на полке: {find_shelf(number)}")
            else:
                print("Документ не найден в базе.")
                
        elif user_cmd == "q" or user_cmd == "Q":
            print("До свидания")
            break
        
        else:
            print("Такой команды не пока что не предусмотрено!")
main()
        
```
